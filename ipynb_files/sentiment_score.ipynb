```python
comments_raw = pd.read_table('./data/artists_comments.csv', delimiter=';', index_col=0)
comments_raw.head(10)
```

```python
# We apply the function split to the column comments
comments = pd.DataFrame({'artist_id': comments_raw['artist_id'], 'comments': comments_raw['comments'].apply(lambda row: row.split('\t'))})

# We apply the function strip to the column comments and remove all the 'NA'
comments['comments'] = comments['comments'].apply(lambda x: [comment.strip() for comment in x if comment != 'NA'])

# Now we have a table with all the comments of all artists
comments
```

```python
#comments['comments'] = comments['comments'].apply(lambda x: [GoogleTranslator(source='auto', target='en').translate(com) for com in x])

translator = Translator()

for i, comment in enumerate(comments['comments']):
    try:
        # Translate the comment to English
        translated_comment = translator.translate(comment, src='auto', dest='en').text
        # Update the comment in the DataFrame with the translated one
        comments.at[i, 'comments'] = translated_comment
    except Exception as e:
        # Print the error message if translation fails
        print(f"Error translating comment {i}: {e}")
    print(i)
```
