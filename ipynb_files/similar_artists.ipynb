```python
# We import the data we have created with scrapping.R
artists = pd.read_csv('./data/artists_type.csv', delimiter=';', index_col=0)
artists.head(10)
```

```python
# We import the data
user = pd.read_csv('./data/user_artists_gp6.dat', delimiter='\t').sort_values('userID')
user[user['userID'] == 505]
```

```python
def get_similar_artists(userid, min_occurence=2, top=5):
    """
    This function takes an user id and return the artists that are similar
    to the one that the user listen to. 

    Arg:
        arg1: id of the user
        arg2: minimum occurence to be accepted as a similar artist
        arg3: the number of top listened artists by the user we take to get the tag
    
    Return:
        the Serie of similar artists and their occurence
    """

    # We get all the rows that contain the userid and we keep only the artistID
    v_artists_listened = user[user['userID'] == int(userid)].sort_values('weight', ascending=False)['artistID'].head(top).to_list()   
    
    # We get only the rows of 'artsts' that are in 'v_reco' and drop all the columns except the music tags.
    artists_id = pd.DataFrame(artists[artists['artist_id'].isin(v_artists_listened)])

    # We take the values of all the columns in a pandas Series
    tag_id = pd.Series(artists_id.iloc[:, 2:5].values.flatten())

    # We drop the id : 0
    tag_id = tag_id[tag_id != 0]

    # We get the occurence of the id's
    tag_id = tag_id.value_counts()

    # We return tag_id
    return tag_id[tag_id >= min_occurence]
```

```python
for i in range(2, 300):
    userid = i

    if int(userid) in user['userID']:
        sim = get_similar_artists(userid, 4, 10)
        if len(sim) > 0:
            print("ID:", userid, "Similar artists to listen : ", sim)
    else:
        print("userID not in database")
# We keep all artists the user listen
```
